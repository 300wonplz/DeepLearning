```python
import matplotlib.pyplot as plt
import numpy as np
```

## State base

```python
x = [1,1, 5, 7, 11, 12, 19]
y = [11, 10, 7, 5, 4, 1, 0]

plt.figure(figsize=(15, 3))
plt.plot(x, y)
plt.xlim(1, 10)
plt.ylim(-3, 20)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Line Plot')
plt.suptitle('Figure Title', size=20, y=1.03)
```

## OO base

```python
fig, ax = plt.subplots(figsize=(15,3))
ax.plot(x, y)
ax.set_xlim(1, 10)
ax.set_ylim(-3, 20)
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_title('Line Plot')
fig.suptitle('Figure Title', size=20, y=1.03)
```

```python
fig, ax = plt.subplots(2, 4, figsize=(15,6))
ax[0][0].plot(x,y, linestyle='--', color='.5', label='test')
ax[0][0].legend('upper right')

ax[0][2].scatter(x,y)
ax[1][1].bar(x,y)
ax[1][2].fill(x, y, 'g', alpha=0.5)
ax[1][3].hist(x, bins=15)
```

```python
import matplotlib.mlab as mlab
import math

mu = 0
variance = 1
sigma = math.sqrt(variance)
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
y = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
plt.fill(x,mlab.normpdf(x, mu, 1), alpha=0.5)
plt.fill(x,mlab.normpdf(x, mu, 0.5), alpha=0.5)
plt.show()
```
